#### Data Cleaning

Remove leading dots (..)

````python
detailed_df.loc[:,"Tourism consumption products"]=detailed_df.loc[:,"Tourism consumption products"].str.replace(r"\.\.","",regex=True)
````

````python
detailed_df
````

To be safe, convert number figures to numeric.

What is ```apply()```

````python
# convert to number
detailed_df.loc[:,"2000":"2018"]=detailed_df.loc[:,"2000":"2018"].apply(pd.to_numeric)
````

````python
detailed_df.columns[1:]
````

We noticed that the products are not sorted according to amounts. For visualization purposes, it may be more meaningful to present products according to amounts in descending order.

````python
# sort
# note: ascending=False means descending order

detailed_df.sort_values(by=['2018'],ascending=False)
````

#### Visualize the data

````python
import matplotlib.pyplot as plt

# don't be afraid to copy and paste from past work
fig, ax = plt.subplots(1)

title_font = {"fontfamily":"Lato", "fontsize":24}
label_font = {"fontfamily":"Lato", "fontsize":12}

ax.set_title("Tourism Consumption Products (2018)", **title_font)

# Review how bar plots can be made
ax.bar(height=detailed_df.sort_values(by=['2018'],ascending=False).loc[:,"2018"], x=detailed_df.sort_values(by=['2018'],ascending=False).loc[:,"Tourism consumption products"], color="Green")
ax.set_ylabel("Spending (in Millions of PHP)", **label_font)
ax.set_xlabel("Product", **label_font)

for tick in ax.get_xticklabels():
    tick.set_rotation(90)

````

## Network Analysis

To do in class: Discuss basic graph theory, nodes and edges

From [Wikipedia](https://en.wikipedia.org/wiki/Graph_theory):

In mathematics, graph theory is the study of graphs, which are mathematical structures used to model pairwise relations between objects. A graph in this context is made up of vertices (also called nodes or points) which are connected by edges (also called links or lines). A distinction is made between undirected graphs, where edges link two vertices symmetrically, and directed graphs, where edges link two vertices asymmetrically; see Graph (discrete mathematics) for more detailed definitions and for other variations in the types of graph that are commonly considered. Graphs are one of the prime objects of study in discrete mathematics.

In one more general sense of the term allowing multiple edges,[3][4] a graph is an ordered triple G = (V, E, ) comprising:

V a set of vertices (also called nodes or points);
E a set of edges (also called links or lines);

````python
import networkx as nx
````

````python
G=nx.Graph()
````

````python
G
````

````python
G.add_node("@jobenilagan")
````

````python
G.add_nodes_from(["@joeilagan","@eeebieee"])
````

````python
G.add_node("@chums")
````

````python
G
````

#### Add edges

````python
G.add_edge("@jobenilagan","@joeilagan",reltype="follower")
````

````python
e=("@jobenilagan","@eeebieee")
G.add_edge(*e,reltype="follower")
````

````python
G.add_edges_from([("@chums","@jobenilagan"),("@joeilagan","@jobenilagan")])
````

#### Graph Statistics

````python
G.number_of_nodes()
````

````python
G.number_of_edges()
````
